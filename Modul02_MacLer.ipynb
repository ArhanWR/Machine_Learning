```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = '/content/wbc.csv'
dp = pd.read_csv(data)
dp.head()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
dp = dp[['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'texture_worst', 'perimeter_worst', 'area_worst']]
dp.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
lep = LabelEncoder() # membuat objek dari LabelEncoder
dp['diagnosis'] = lep.fit_transform(dp['diagnosis']) # proses encoding
dp.head()
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
dp['radius_mean'] = std.fit_transform(dp[['radius_mean']])
dp['texture_mean'] = std.fit_transform(dp[['texture_mean']])
dp['perimeter_mean'] = std.fit_transform(dp[['perimeter_mean']])
dp['area_mean'] = std.fit_transform(dp[['area_mean']])
dp['texture_worst'] = std.fit_transform(dp[['texture_worst']])
dp['perimeter_worst'] = std.fit_transform(dp[['perimeter_worst']])
dp['area_worst'] = std.fit_transform(dp[['area_worst']])
dp.head()
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import pandas as pd
dp = pd.read_csv('/content/wbc.csv')
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 80:20
dp_train, dp_unseen = train_test_split(dp, test_size=0.2, random_state=0, stratify=dp['diagnosis'])

# Split lagi antara validasi dan testing
dp_val, dp_test = train_test_split(dp_unseen, test_size=0.2, random_state=0, stratify=dp_unseen['diagnosis'])

# Cek masing-masing ukuran data
print(f'Jumlah label data asli:\n{dp.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{dp_train.diagnosis.value_counts()}')
print(f'Jumlah label data val:\n{dp_val.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{dp_test.diagnosis.value_counts()}')
```
